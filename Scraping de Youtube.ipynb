```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd

urlVideo = 'https://www.youtube.com/feeds/videos.xml?channel_id=UCdulIs-x_xrRd1ezwJZR9ww'


r = requests.get(urlVideo, headers = {"User-Agent": "Chrome/50.0.2661.94"})
data = r.content

root = ET.fromstring(data.decode())

lista = root.findall('.//')

title = []
views = []
href = []
count = []
name = []
description = []

for item in lista:
    titulo = item.find('{http://search.yahoo.com/mrss/}title')
    if titulo is None:
        pass
    else:
        print('\nTitulo: ', titulo.text)
        title.append(titulo.text)
    
    visto = item.get('views')
    if visto is None:
        pass
    else:
        print('Visualizaciones: ', visto)
        views.append(visto)
    
    url = item.get("href")
    if url is None:
        pass
    else:
        print("Video: ", url)
        href.append(url)
        
    like = item.get('count')
    if like is None:
        pass
    else:
        print('Me gusta: ', like)
        count.append(like)
        
    descripcion = item.find('{http://search.yahoo.com/mrss/}description')
    if descripcion is None:
        pass
    else:
        print('Descripción: ', descripcion.text)
        description.append(descripcion.text)
        
    autor = item.find('{http://www.w3.org/2005/Atom}name')
    if autor is None:
        pass
    else:
        print('Nombre del autor: ', autor.text)
        name.append(autor.text)
        
    
```

```python
df = pd.DataFrame({"Titulo":title,
                   "Visualizaciones":views, 
                    "Me Gusta":count,
                    "Descripcion del video":description
                  
})
```
